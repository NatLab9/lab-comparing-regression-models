```python
data =  pd.read_csv('DATA.csv')
data.head()
```

```python
#1. In this final lab, we will model our data. Import sklearn train_test_split and separate the data.
import pandas as pd
from sklearn import linear_model
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
```

```python
X = data.drop('total_claim_amount',axis=1)
y = data['total_claim_amount']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=100)
```

```python
#2. Try a simple linear regression with all the data to see whether we are getting good results.
lm = linear_model.LinearRegression()
model = lm.fit(X_train,y_train)
predictions = lm.predict(X_test)
r2_score(y_test, predictions)

#The 85,5% is pretty accurate 
```

```python
#3. Great! Now define a function that takes a list of models and train (and tests) them so we can try a lot of them without repeating code.

import numpy as np
from sklearn.metrics import r2_score
from sklearn.neural_network import MLPRegressor

def train_and_test_models(models, X_train, y_train, X_test, y_test):
    trained_models = []
    test_scores = []
    
    for model in models:
        # Train the model
        model.fit(X_train, y_train)
        
        # Test the model
        y_pred = model.predict(X_test)
        score = r2_score(y_test, y_pred)
        
        # Store the trained model and test score
        trained_models.append(model)
        test_scores.append(score)
        
    return trained_models, test_scores
```

```python
#4. Use the function to check LinearRegressor and KNeighborsRegressor.
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import train_test_split


models = [LinearRegression(), KNeighborsRegressor(), MLPRegressor(hidden_layer_sizes=(100,))]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=100)

trained_models, test_scores = train_and_test_models(models, X_train, y_train, X_test, y_test)

for i, model in enumerate(trained_models):
    print(f"Model {i}: {type(model).__name__} (Test score: {test_scores[i]:.2f})")
```
